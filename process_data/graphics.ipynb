```python
import json
import pandas as pd
import skimage.measure
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('agile_acceleration.json') as fp:
    agile_acceleration = pd.Series(json.load(fp))
    
with open('agile_velocity.json') as fp:
    agile_velocity = pd.Series(json.load(fp))
    
with open('fast_acceleration.json') as fp:
    fast_acceleration = pd.Series(json.load(fp))

with open('fast_velocity.json') as fp:
    fast_velocity = pd.Series(json.load(fp))
```

```python
suffix = 'compressed_normalized'

with open(f'agile_acceleration_{suffix}.json') as fp:
    agile_acceleration = pd.Series(json.load(fp))
    
with open(f'agile_velocity_{suffix}.json') as fp:
    agile_velocity = pd.Series(json.load(fp))

with open(f'fast_acceleration_{suffix}.json') as fp:
    fast_acceleration = pd.Series(json.load(fp))

with open(f'fast_velocity_{suffix}.json') as fp:
    fast_velocity = pd.Series(json.load(fp))
```

```python
suffix = 'compressed_normalized'

with open(f'agile_acceleration_{suffix}.json', 'w') as fp:
    json.dump(agile_acceleration.tolist(), fp, indent=2)
    
with open(f'agile_velocity_{suffix}.json', 'w') as fp:
    json.dump(agile_velocity.tolist(), fp, indent=2)
    
with open(f'fast_acceleration_{suffix}.json', 'w') as fp:
    json.dump(fast_acceleration.tolist(), fp, indent=2)

with open(f'fast_velocity_{suffix}.json', 'w') as fp:
    json.dump(fast_velocity.tolist(), fp, indent=2)
```
