# FAST PLANNER

```python
from bagpy import bagreader

import pandas as pd
import matplotlib.pyplot as plt
```

```python
def normalize(x, y, z):
    a = (x ** 2 + y ** 2 + z ** 2) ** 0.5
    return a / a.abs().max()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
bag = bagreader('2022-05-17-14-36-08.bag')
```

```python
bag.topic_table
```

```python
df = pd.read_csv(bag.message_by_topic('/planning/pos_cmd'))
```

```python
df.columns
```

```python
x = df['velocity.x']
y = df['velocity.y']
z = df['velocity.z']
_v = normalize(x, y, z)
v = _v[abs(_v) > 10e-6]

# t = df['Time']
# t = _t[abs(_v) > 10e-6]
# t = t - t.min()

import json
with open('fast_velocity.json', 'w') as fp:
    json.dump(v.dropna().tolist(), fp)
    
plt.figure(figsize=(8, 8))
plt.grid(True, 'both')
plt.plot(v, label="normalized velocity", color='black')
plt.legend(fontsize="x-large")
plt.show()
```

```python
x = df['acceleration.x']
y = df['acceleration.y']
z = df['acceleration.z']
_a = normalize(x, y, z)
a = _a[abs(_v) > 10e-6].rolling(window=11).mean()

# _t = df['Time']
# t = _t[abs(_v) > 10e-6]
# t = t - t.min()

import json
with open('fast_acceleration.json', 'w') as fp:
    json.dump(a.dropna().tolist(), fp)

plt.figure(figsize=(8, 8))
plt.grid(True, 'both')
plt.plot(a, label="normalized acceleration", color='black')
plt.legend(fontsize="x-large")
plt.show()
```
